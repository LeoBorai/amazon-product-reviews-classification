Lets ease a bit the detection to allow either title or text to be English in order to determine wether the language is English, this could affect precision but might also bring in more samples to train the classification model.

```python
from langdetect import detect

def detect_english(title, text):
    if detect(title) == 'en' or detect(text) == 'en':
        return 'en'
    return 'unknown'

df['lang'] = df.apply(lambda x: detect_english(x.title, x.text), axis=1)
```

```python
df['lang'].value_counts()
```

This change brough in almost 98% of our samples! Now other items which are not English could be removed.

```python
df = df.query('lang == "en"')
df['lang'].value_counts()
```

## Transforming Data

Prepare data for trainning, the dataset is splitted using Pareto Principle (80/20), then data is transformed to satisfy algorithm requirements.

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

reviews = df['text'].values
ratings = df['rating'].values

reviews_train, reviews_test, y_train, y_test = train_test_split(reviews,
                                                               ratings, test_size=0.2, random_state=1000)

vectorizer = CountVectorizer()
vectorizer.fit(reviews_train)
x_train = vectorizer.transform(reviews_train)
x_test = vectorizer.transform(reviews_test)
```

```python
# Train Model
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
classifier.fit(x_train, y_train)
```

Should extend max iterations (default to 100) to achieve covergence, the algorithm used behind
scenes is "Limited-memory Broydenâ€“Fletcherâ€“Goldfarbâ€“Shanno Algorithm" which performs better in
smaller datasets. Extending maximum interations will NOT GUARANTEE COVERGENCE but might help.

```python
classifier = LogisticRegression(solver='lbfgs', max_iter=1000)
classifier.fit(x_train, y_train)
```

Store Model into Pickle for Easier Retrieval

```python
import pickle

pickle.dump(classifier, open('model.pkl', 'wb'))
```

### Testing Model

Using the remaining `x_test` dataset we can test our model.

```python
import numpy as np

predicted = classifier.predict(x_test)
accurancy = np.mean(predicted == y_test)

print('Accurancy: ', round(accurancy,2))
```

Apply _Confusion Matrix_ metric on model to determine wether the model could mislabel classes.

```python
from sklearn import metrics

RATINGS = [1, 2, 3, 4, 5]

print(metrics.confusion_matrix(y_test, predicted, labels = RATINGS))
```

The main diagonal compound by:

{1, 2, 16, 15, 124}

Shows that the model made 124 correct predictions for 5 star reviews and dropped to 15 correct predictions for 4 star reviews, predictions for 3 star reviews are also closer to 4 star review performance.

For 1 star and 2 stars it drops to 1 and 2 respectively.

Predictions on 5 star reviews outstands other reviews, this might be due to a higher presence of 5 stars reviews in the dataset than for other classes (4 and others).

```python
print(metrics.classification_report(y_test, predicted, labels = RATINGS))
```

The support metric shows the number of samples per class of ratings, showing that for 5 star ratings we have 144 while we have 50 for 4 stars and drops event more on subsequent classes.
